```python
from datetime import datetime

import kagglehub
path = kagglehub.dataset_download("elvinrustam/books-dataset")
print("Path to dataset files:", path)
```

```python
import pandas as pd
```

```python
books = pd.read_csv(f"{path}/BooksDataset.csv")
```

```python
books
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ax = plt.axes()
sns.heatmap(books.isna().transpose(), ax=ax, cbar=False)
plt.xlabel("Columns")
plt.ylabel("Missing values")
plt.show()
```

```python
import numpy as np
```

```python
books['missing_description'] = np.where(books['Description'].isna(),1,0)
```

```python
import datetime
```

```python
books['Publish Date'][0]
```

```python
books['age_of_book'] = books['Publish Date'].apply(lambda x: 2025-int(x.split(' ')[-1].strip()))
```

```python
books = books[books['age_of_book']>=0]
```

```python
books_missing = books[~books['Description'].isna()]
```

```python
books_missing
```

```python
books_missing["Category"].value_counts().reset_index().sort_values("count", ascending=False)
```

```python
books_missing["words_in_description"] = books_missing["Description"].str.split().str.len()
```

```python
books_missing["words_in_description"]
```

```python
books_missing.loc[books_missing["words_in_description"].between(1,4),"Description"]
```

```python
books_missing.loc[books_missing["words_in_description"].between(15,24),"Description"]
```

```python
books_missing_25_words = books_missing[books_missing["words_in_description"]>=25]
```
